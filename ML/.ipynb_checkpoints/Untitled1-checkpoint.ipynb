```python
import numpy as np
import pandas as pd
import matplotlib.pylab as plt
%matplotlib inline
from matplotlib.pylab import rcParams
#from datetime import datetime
import datetime
rcParams['figure.figsize']= 20,12

dataset = pd.read_csv("new_data.csv")

#dataset['OnlyDate'] = pd.to_datetime(dataset['Date_and_time'], infer_datetime_format = True).dt.year
dataset['OnlyDate'] = pd.to_datetime(dataset['Date_and_time'], infer_datetime_format = True)
dataset['OnlyDate'] = dataset['OnlyDate'].apply(lambda dataset : datetime.datetime(year = dataset.year, month = dataset.month, day = dataset.day))
#dataset['Date_and_time'].year
#dataset['OnlyDate'] = dataset['OnlyDate'].map(lambda x: 100*x.year + x.month)
#dataset['OnlyDate'] = dataset.date_column.dt.to_period('M')
dataset = dataset[['OnlyDate','Number_of_patients']]
indexedDataset = dataset.set_index(['OnlyDate'])
indexedDataset = indexedDataset.groupby(['OnlyDate']).sum()
#print(indexedDataset.head(5))

```

```python
indexedDataset.head(10)
```

```python
plt.xlabel("Month")
plt.ylabel("Number_of_patients")
plt.plot(indexedDataset)
```

```python
rolmean = indexedDataset.rolling(window = 365).mean()
rolstd = indexedDataset.rolling(window = 365).std()
print(rolmean,rolstd)
```

```python
orig = plt.plot(indexedDataset, color = 'blue', label = 'Original')
mean = plt.plot(rolmean, color = 'red', label = 'Rolling Mean')
std = plt.plot(rolstd, color = 'black', label = 'Rolling Standard')
plt.legend(loc='best')
plt.title('Rolling Mean and Standard Deviation')
plt.show(block=False)
```

```python
from statsmodels.tsa.stattools import adfuller
print('Results')
dftest = adfuller(indexedDataset['Number_of_patients'], autolag = 'AIC')
dfoutput = pd.Series(dftest[0:4], index = ['Test Statistics', 'p-value', 'Lags Used', 'Number ofObservation Used'])
for key,value in dftest[4].items():
    dfoutput['Critical value (%s)' %key] = value
print(dfoutput)
```

```python
indexedDataset_logScale = np.log(indexedDataset)
plt.plot(indexedDataset_logScale)
```

```python
movingAverage = indexedDataset_logScale.rolling(window=365).mean()
movingSTD = indexedDataset_logScale.rolling(window=365).std()
plt.plot(indexedDataset_logScale)
plt.plot(movingAverage, color = 'red')
```

```python
datasetLogScaleMinusMovingAverage = indexedDataset
datasetLogScaleMinusMovingAverage.head(12)

datasetLogScaleMinusMovingAverage.dropna(inplace = True)
datasetLogScaleMinusMovingAverage.head(10)
```

```python
from statsmodels.tsa.stattools import adfuller
def test_stationarity(timeseries):
    movingAverage = timeseries.rolling(window = 365).mean()
    movingSTD = timeseries.rolling(window = 365).std()
    print(movingAverage,movingSTD)
    
    orig = plt.plot(timeseries, color = 'blue', label = 'Original')
    mean = plt.plot(movingAverage, color = 'red', label = 'Rolling Mean')
    std = plt.plot(movingSTD, color = 'black', label = 'Rolling Standard')
    plt.legend(loc='best')
    plt.title('Rolling Mean and Standard Deviation')
    plt.show(block=False)

    print('DF Test')
    
    dftest = adfuller(timeseries['Number_of_patients'], autolag = 'AIC')
    dfoutput = pd.Series(dftest[0:4], index = ['Test Statistics', 'p-value', 'Lags Used', 'Number ofObservation Used'])
    for key,value in dftest[4].items():
        dfoutput['Critical value (%s)' %key] = value
    print(dfoutput)
```

```python
test_stationarity(datasetLogScaleMinusMovingAverage)
```

```python
exponentialDecayWeightedAverage = indexedDataset_logScale.ewm(halflife= 365, min_periods = 0 , adjust=True).mean()
plt.plot(indexedDataset_logScale)
plt.plot(exponentialDecayWeightedAverage,color = 'red')
```

```python
datasetLogScaleMinusMovingExponentialDecayAverage = indexedDataset_logScale - exponentialDecayWeightedAverage
test_stationarity(datasetLogScaleMinusMovingExponentialDecayAverage)
```

```python
datasetLogDiffShifting = indexedDataset_logScale - indexedDataset_logScale.shift()
plt.plot(datasetLogDiffShifting)
```

```python
datasetLogDiffShifting.dropna(inplace=True)
test_stationarity(datasetLogDiffShifting)
```

```python
indexedDataset_logScale.dropna(inplace=True)
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose
decomposition = seasonal_decompose(indexedDataset_logScale)
trend = decomposition.trend
seasonal = decomposition.seasonal
residual = decomposition.resid

plt.subplot(411)
plt.plot(indexedDataset_logScale, label='Original')
plt.legend(loc='best')
plt.subplot(412)
plt.plot(trend, label='Trend')
plt.legend(loc='best')
plt.subplot(413)
plt.plot(seasonal, label='seasonal')
plt.legend(loc='best')
plt.subplot(414)
plt.plot(residual, label='residual')
plt.legend(loc='best')

plt.tight_layout()

decomposedLogData = residual
decomposedLogData.dropna(inplace = True)
test_stationarity(decomposedLogData)
```
