```python
import numpy as np
import pandas as pd
import matplotlib.pylab as plt
%matplotlib inline
from matplotlib.pylab import rcParams
#from datetime import datetime
import datetime
rcParams['figure.figsize']= 10,6

dataset = pd.read_csv("new_data.csv")

#dataset['OnlyDate'] = pd.to_datetime(dataset['Date_and_time'], infer_datetime_format = True).dt.year
dataset['OnlyDate'] = pd.to_datetime(dataset['Date_and_time'], infer_datetime_format = True)
dataset['OnlyDate'] = dataset['OnlyDate'].apply(lambda dataset : datetime.datetime(year = dataset.year, month = dataset.month, day = int('01')))
#dataset['Date_and_time'].year
#dataset['OnlyDate'] = dataset['OnlyDate'].map(lambda x: 100*x.year + x.month)
#dataset['OnlyDate'] = dataset.date_column.dt.to_period('M')
dataset = dataset[['OnlyDate','Number_of_patients']]
indexedDataset = dataset.set_index(['OnlyDate'])
indexedDataset = indexedDataset.groupby(['OnlyDate']).sum()
#print(indexedDataset.head(5))

```

```python
indexedDataset.head(10)
```

```python
plt.xlabel("Month")
plt.ylabel("Number_of_patients")
plt.plot(indexedDataset)
```

```python
rolmean = indexedDataset.rolling(window = 12).mean()
rolstd = indexedDataset.rolling(window = 12).std()
print(rolmean,rolstd)
```

```python
orig = plt.plot(indexedDataset, color = 'blue', label = 'Original')
mean = plt.plot(rolmean, color = 'red', label = 'Rolling Mean')
std = plt.plot(rolstd, color = 'black', label = 'Rolling Standard')
plt.legend(loc='best')
plt.title('Rolling Mean and Standard Deviation')
plt.show(block=False)
```

```python
from statsmodels.tsa.stattools import adfuller
print('Results')
dftest = adfuller(indexedDataset['Number_of_patients'], autolag = 'AIC')
dfoutput = pd.Series(dftest[0:4], index = ['Test Statistics', 'p-value', 'Lags Used', 'Number ofObservation Used'])
for key,value in dftest[4].items():
    dfoutput['Critical value (%s)' %key] = value
print(dfoutput)
```
